```sql
-- First and Last Order Date Per Customer

WITH orders as (SELECT 
cd.customer_unique_id as cust_id,
CAST(od.order_approved_at as DATE) as order_date
FROM olist_orders_dataset od
JOIN olist_customers_dataset cd on cd.customer_id = od.customer_id
WHERE od.order_status NOT IN ('canceled','unavailable'))

SELECT cust_id,
min(order_date) as first_order,
CASE WHEN COUNT(*) > 1 THEN max(order_date)
ELSE NULL END AS last_order
FROM orders
GROUP BY cust_id
ORDER BY last_order DESC

```
