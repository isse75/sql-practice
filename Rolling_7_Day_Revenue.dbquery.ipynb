```sql
-- Rolling 7 day Revenue

WITH Daily as (SELECT CAST(od.order_purchase_timestamp as date) as purchase_day, 
ROUND(sum(oi.price), 2) as daily_revenue
FROM olist_orders_dataset od
JOIN olist_order_items_dataset oi ON od.order_id = oi.order_id
GROUP BY purchase_day)

SELECT *,
SUM(daily_revenue) OVER (ORDER BY purchase_day
RANGE BETWEEN INTERVAL 6 DAY PRECEDING AND CURRENT ROW) as rolling_7_day_revenue
FROM Daily
ORDER BY purchase_day
```
