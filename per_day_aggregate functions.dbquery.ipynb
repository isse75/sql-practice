```sql
-- For each order date, produce:

-- - `order_date`
-- - `orders_total`
-- - `orders_delivered`

-- Where `orders_total` is the count of all orders on that date and `orders_delivered` is the count where `order_status` is delivered.

SELECT 
order_purchase_timestamp::date as order_date,
count(order_id) as orders_total,
count(case when order_status = 'delivered' then order_id end) as orders_delivered
FROM olist_orders_dataset
GROUP BY order_date
ORDER BY order_date asc
```
