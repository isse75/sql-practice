```sql
-- Create a query that:

-- - converts `order_purchase_timestamp` to a proper timestamp column `order_ts`
-- - derives `order_date` as the date part

SELECT 
    CAST(order_purchase_timestamp AS TIMESTAMP) AS order_ts,
    CAST(LEFT(order_purchase_timestamp, 10) AS DATE) AS order_date
FROM olist_orders_dataset
LIMIT 50

```
